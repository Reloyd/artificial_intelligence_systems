### Задание 2

Реализуйте с нуля алгоритм линено регрессии  без использования регуляризации.

```python
class LinearRegression:
    def __init__(self):
        self.coef_ = None  # Коэффициенты b (влияние признаков)
        self.intercept_ = None  # a (начало линии)
        self.is_fitted = False

    def fit(self, X, y):
        # X — признаки (таблица чисел), y — цены
        # Добавляем столбец 1 для intercept
        X_with_ones = np.column_stack([np.ones(len(X)), X])
        
        # Формула: коэффициенты = (X^T * X)^(-1) * X^T * y
        XtX = X_with_ones.T @ X_with_ones
        XtX_inv = np.linalg.inv(XtX)
        Xty = X_with_ones.T @ y
        
        beta = XtX_inv @ Xty  # Все коэффициенты вместе
        self.intercept_ = beta[0]  # Первое — intercept
        self.coef_ = beta[1:]  # Остальные — для признаков
        self.is_fitted = True
        
        print(f"Обучено! Intercept: {self.intercept_:.0f}, Коэффициенты: {self.coef_}")

    def predict(self, X):
        # Предсказать цены: a + b1*x1 + b2*x2 + ...
        if not self.is_fitted:
            print("Сначала обучите модель!")
            return None
        X_with_ones = np.column_stack([np.ones(len(X)), X])
        return X_with_ones @ np.hstack([self.intercept_, self.coef_])

    def score(self, X, y):
        # Точность: R² (от 0 до 1, ближе к 1 — лучше)
        y_pred = self.predict(X)
        ss_res = np.sum((y - y_pred)**2)  # Ошибка
        ss_tot = np.sum((y - np.mean(y))**2)  # Общая разбросанность
        return 1 - (ss_res / ss_tot)
```

### Задание 3

Реализуйте с нуля алгоритм линейной регрессии для многоклассовой классификации с L1-регуляризацией.

```python
class LinearRegressionL1:
    def __init__(self, learning_rate=0.001, n_iterations=1000, l1_lambda=0.01):
        """
        Линейная регрессия с L1 (Lasso). Градиентный спуск.
        learning_rate: шаг (маленький для стабильности).
        n_iterations: итераций.
        l1_lambda: штраф L1 (0 — без регуляризации).
        """
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.l1_lambda = l1_lambda
        self.coef_ = None  # Коэффициенты (b)
        self.intercept_ = None  # Перехват (a)
        self.is_fitted = False
        # Для нормализации
        self.X_mean_ = None
        self.X_std_ = None
        self.y_mean_ = None
        self.y_std_ = None

    def fit(self, X, y):
        """
        Обучение. Автоматическая нормализация + градиентный спуск.
        X: признаки, y: Price_num.
        """
        # 1. Приведение к float64 (исправление dtype ошибки)
        X = X.astype(np.float64)
        y = y.astype(np.float64)
        
        n_samples, n_features = X.shape
        
        # 2. Нормализация (обязательно, иначе NaN!)
        self.X_mean_ = np.mean(X, axis=0)
        self.X_std_ = np.std(X, axis=0) + 1e-8  # +epsilon от деления на 0
        X_norm = (X - self.X_mean_) / self.X_std_
        
        self.y_mean_ = np.mean(y)
        self.y_std_ = np.std(y) + 1e-8
        y_norm = (y - self.y_mean_) / self.y_std_
        
        # 3. Добавляем столбец 1 для intercept (не нормализуем)
        X_with_bias = np.column_stack([np.ones(n_samples), X_norm])
        n_total = X_with_bias.shape[1]
        
        # 4. Инициализация весов нулями (стабильно)
        weights = np.zeros(n_total)
        
        for iteration in range(self.n_iterations):
            # Предсказания на нормализованных
            y_pred = X_with_bias @ weights
            error = y_pred - y_norm
            
            # MSE
            mse = np.mean(error ** 2)
            if np.isnan(mse) or np.isinf(mse):
                print(f"Ошибка NaN на итерации {iteration}! Уменьшите learning_rate.")
                break
            
            # Градиент MSE + L1 (только для coef, не bias)
            grad = (2 / n_samples) * (X_with_bias.T @ error)
            grad[1:] += self.l1_lambda * np.sign(weights[1:])  # L1: |w|
            
            # Обновление (маленький шаг)
            weights -= self.learning_rate * grad
            
            # L1: обнуляем слабые веса
            weights[1:][np.abs(weights[1:]) < 1e-5] = 0
            
            if iteration % 200 == 0:
                print(f"Итерация {iteration}, MSE: {mse:.4f}")
        
        # 5. Денормализация весов обратно к оригинальным единицам
        self.intercept_ = weights[0] * self.y_std_ + self.y_mean_
        self.coef_ = (weights[1:] * self.y_std_) / self.X_std_
        self.is_fitted = True
        
        print(f"Обучено! Intercept: {self.intercept_:.0f} руб.")
        print(f"Ненулевых весов: {np.sum(np.abs(self.coef_) > 1e-5)} из {n_features}")
        return self

    def predict(self, X):
        """Предсказания. Автоматическая нормализация."""
        if not self.is_fitted:
            print("Сначала обучите fit()!")
            return None
        
        X = X.astype(np.float64)
        n_samples = X.shape[0]
        
        # Нормализация (как при обучении)
        X_norm = (X - self.X_mean_) / (self.X_std_ + 1e-8)
        X_with_bias = np.column_stack([np.ones(n_samples), X_norm])
        
        # Предсказания на нормализованных + денормализация
        y_pred_norm = X_with_bias @ np.hstack([
            (self.intercept_ - self.y_mean_) / self.y_std_, 
            self.coef_ * self.X_std_ / self.y_std_
        ])
        return y_pred_norm * self.y_std_ + self.y_mean_

    def score(self, X, y):
        """R² метрика (0-1, 1=идеально)."""
        y_pred = self.predict(X)
        ss_res = np.sum((y - y_pred) ** 2)
        ss_tot = np.sum((y - np.mean(y)) ** 2)
        r2 = 1 - (ss_res / ss_tot)
        return r2
```
